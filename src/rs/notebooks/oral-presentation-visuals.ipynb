```python
import pylab
import numpy as N
f = 10.
w = 2. * N.pi * f
time_interval = 1
fig = pylab.figure()
for i, samples in 500:
    pylab.subplot(2, 2, i+1)
    pylab.title('%i samples'%samples)
    t = N.linspace(0, time_interval, samples)
    y = N.sin(w * t)
    pylab.plot(t, y, '.-')
fig.show()
```
